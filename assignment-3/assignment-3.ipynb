```python
import numpy as np
import pandas as pd

from sklearn import (
    model_selection, linear_model, metrics, 
    neighbors, naive_bayes,
)
```

# weakest link classification

## loading the data

```python
X = 'AimoScore_WeakLink_big_scores (1).xls'
Y = '20190108 scores_and_weak_links.xlsx'

X = pd.read_excel(X, index_col= 'ID').drop(['AimoScore','EstimatedScore'], axis= 1)
Y = pd.read_excel(Y, index_col= 'ID').drop(['Date'     , 'SCORE'        ], axis= 1).T.apply(np.argmax).rename('weakest link')

df = X.join(Y)
df
```

## applying train test split

```python
X = df.drop('weakest link', axis= 1)
Y = df['weakest link']

trainX, testX, trainY, testY = model_selection.train_test_split(X, Y, test_size= 0.2)
```

## training models

### Nearest neighbours

```python
model = neighbors.KNeighborsClassifier(n_neighbors= 10)

model.fit(trainX,trainY)

testP = model.predict(testX)
metrics.accuracy_score(testY,testP)
```

### Naive Bayes

```python
model = naive_bayes.GaussianNB()

model.fit(trainX,trainY)

testP = model.predict(testX)
metrics.accuracy_score(testY,testP)
```

### logistic regressor

```python
model = linear_model.LogisticRegression()

model.fit(trainX,trainY)

testP = model.predict(testX)
metrics.accuracy_score(testY,testP)
```
